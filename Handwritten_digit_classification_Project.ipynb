### Create a neural network

### Preprocess xtrain, ytrain, xtest, ytest

```python
xtrain_scl = xtrain.astype('float32')/255
xtest_scl = xtest.astype('float32')/255
```

### Convert y to categorical

```python
from keras.utils import to_categorical
ytrain_cat = to_categorical(ytrain)
ytest_cat = to_categorical(ytest)
```

```python
ytrain[0:5]
```

```python
ytrain_cat[0:5]
```

image.png

```python
from keras.models import Sequential
from keras.layers import Input, Conv2D, MaxPool2D, Flatten, Dropout, Dense
```

```python
model = Sequential()

# Create a Input layer
model.add(Input(shape=(28, 28, 1)))
# Create a convolution layer 1
model.add(Conv2D(filters=32, kernel_size=(3,3),activation='relu'))
# Create a max pool layer 1
model.add(MaxPool2D(pool_size=(2,2)))
# Create convolution layer 2
model.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
# Create max pool layer 2
model.add(MaxPool2D(pool_size=(2,2)))

# Flatten layer
model.add(Flatten())

# Dropout layer
model.add(Dropout(0.5))

# Dense layer (Multiclass - Softmax)
model.add(Dense(units=10, activation='softmax'))
```

### Compile the model

```python
model.compile(optimizer='adam', loss='categorical_crossentropy',
              metrics=['accuracy'])
```

### Train model for 15 epochs

```python
hist = model.fit(xtrain_scl, ytrain_cat, validation_split=0.1, epochs=15)
```

### Plot the learning curve

```python
plt.figure(figsize=(10, 5))
plt.plot(hist.history['loss'], label='Train Loss')
plt.plot(hist.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Categorical Cross Entropy')
plt.title('Learning Curve')
plt.show()
```

### Evaluate model in train and test

```python
model.evaluate(xtrain_scl, ytrain_cat)
```

```python
model.evaluate(xtest_scl, ytest_cat)
```

### Predict Probability for train and test

```python
yprob_train = model.predict(xtrain_scl)
yprob_test = model.predict(xtest_scl)
```

### Get the predictions from probability

```python
import numpy as np
ypred_train = np.argmax(yprob_train, axis=1)
ypred_test = np.argmax(yprob_test, axis=1)
```
